```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('https://github.com/opendatazurich/opendatazurich.github.io/raw/master/automation/badi-counter/badi_counter.csv', parse_dates=[0])
df
```

```python
df_places = df.groupby(['name', 'id', 'slug']).size().reset_index(name="count").sort_values("count", ascending=False).reset_index(drop=True)
df_places
```

```python
# to use different styles, make sure to reload the default to always get clean results
def use_style(style):
    plt.style.use('default')
    plt.style.use(style)
```

```python
def badi_plot(df, name):
    use_style('fivethirtyeight')
    fig, ax = plt.subplots(figsize=(25,25))
    ax.set_title(name, fontsize=24, fontweight='bold', pad=20);

    df.plot(kind="line", x='date', y='occupancy_count', ax=ax, linewidth=2, label="Belegung")
    df.plot(kind="line", x='date', y='max', ax=ax, linewidth=4, label="Maximale Belegung")
    df.plot(kind="line", x='date', y='waiting_count', ax=ax, linewidth=2, label="Wartende")

    # add tick every 2d
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))

    # show date in Zurich timezone
    formatter = mdates.DateFormatter('%d.%m.%Y')
    ax.xaxis.set_major_formatter(formatter)

    # labels
    ax.set_ylabel('Anzahl Personen')
    ax.set_xlabel('Zeit')

    plt.show()
```

```python
# Create DataFrames
df_mq = df[df.name.str.contains('Mythenquai')].reset_index(drop=True)
df_ul = df[df.name == 'Unterer Letten Flussteil'].reset_index(drop=True)
df_ulns = df[df.name == 'Unterer Letten Plansch- und Nichtschwimmerbecken'].reset_index(drop=True)
df_au = df[df.name == 'Flussbad Au-Höngg'].reset_index(drop=True)
df_ks = df[df.name == 'Seebad Katzensee'].reset_index(drop=True)
```

```python
badi_plot(df_mq, "Strandbad Mythenquai")
badi_plot(df_ul, "Unterer Letten Flussteil")
badi_plot(df_ulns, "Unterer Letten Plansch- und Nichtschwimmerbecken")
badi_plot(df_au, "Flussbad Au-Höngg")
badi_plot(df_ks, "Seebad Katzensee")
```
